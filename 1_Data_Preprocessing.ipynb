```python
import pandas as pd

# Path to your CSV file
csv_file = "../data/NIFTY 50_Historical_PR_01011990to11102024.csv"

# Load dataset into pandas DataFrame
df = pd.read_csv(csv_file)

# Rename columns for clarity
df.rename(columns={'Close':'Closing_Price', 'Volume':'Trade_Volume'}, inplace=True)

# Convert 'Date' column to datetime (dayfirst=True since format is DD/MM/YYYY)
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True)

# Sort data by Date ascending (oldest → newest)
df.sort_values('Date', inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

# Preview first 5 rows
df.head()
```

```python
# Keep data from 2000 onwards
df = df[df['Date'] >= '2000-01-01']
df.reset_index(drop=True, inplace=True)

# Preview
df.head()
```

```python
# Check for missing values
print(df.isnull().sum())

# Fill missing values using previous day
df.ffill(inplace=True)  # forward-fill missing values
```

```python
# Filter data from year 2000 onwards

df = df[df['Date'].dt.year >= 2000]

# Reset index after filtering
df.reset_index(drop=True, inplace=True)

# Check first and last dates
print("First Date:", df['Date'].min())
print("Last Date:", df['Date'].max())

# Check number of rows
print("Total Rows After Filtering:", len(df))

# Preview data
df.head()
```

```python
# Create 50-day Moving Average
df['MA_50'] = df['Closing_Price'].rolling(window=50).mean()

# Create 200-day Moving Average
df['MA_200'] = df['Closing_Price'].rolling(window=200).mean()

# Create Daily Return
df['Daily_Return'] = df['Closing_Price'].pct_change()

# Create Trend column (1 = Up, 0 = Down)
df['Trend'] = (df['Closing_Price'] > df['Closing_Price'].shift(1)).astype(int)

# Check new columns
df.head(10)
```

```python
df.dropna(inplace=True)
df.reset_index(drop=True, inplace=True)

df.head()
```

```python
# Check total rows
total_rows = len(df)
print("Total Rows:", total_rows)

# Define training size (80%)
train_size = int(total_rows * 0.8)

# Split dataset
train_data = df[:train_size]
test_data = df[train_size:]

print("Training Data Size:", len(train_data))
print("Testing Data Size:", len(test_data))

# Check last date of training and first date of testing
print("Last Training Date:", train_data['Date'].iloc[-1])
print("First Testing Date:", test_data['Date'].iloc[0])
```

```python
from sklearn.preprocessing import MinMaxScaler
import numpy as np

# Select only Closing Price for LSTM
train_close = train_data[['Closing_Price']]
test_close = test_data[['Closing_Price']]

# Create scaler object
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit scaler only on training data
train_scaled = scaler.fit_transform(train_close)

# Transform testing data
test_scaled = scaler.transform(test_close)

print("Training Scaled Shape:", train_scaled.shape)
print("Testing Scaled Shape:", test_scaled.shape)
```

```python
# Function to create sequences

def create_sequences(data, time_steps=60):
    X = []
    y = []
    
    for i in range(time_steps, len(data)):
        X.append(data[i-time_steps:i, 0])  # last 60 days
        y.append(data[i, 0])              # next day
        
    return np.array(X), np.array(y)
```

```python
# Create sequences for training data
X_train, y_train = create_sequences(train_scaled, 60)

# Create sequences for testing data
X_test, y_test = create_sequences(test_scaled, 60)

print("X_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
print("X_test shape:", X_test.shape)
print("y_test shape:", y_test.shape)
```

```python
# Reshape for LSTM (3D input)
X_train = X_train.reshape(X_train.shape[0], X_train.shape[1], 1)
X_test = X_test.reshape(X_test.shape[0], X_test.shape[1], 1)

print("Reshaped X_train:", X_train.shape)
print("Reshaped X_test:", X_test.shape)
```

```python
import numpy as np

np.save("X_train.npy", X_train)
np.save("y_train.npy", y_train)
np.save("X_test.npy", X_test)
np.save("y_test.npy", y_test)

print("Files saved successfully")
```
